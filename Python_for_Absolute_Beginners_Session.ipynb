Mini Practice (Combining Everything)

```python
#🔹 Section 5: Mini Practice (Combining Everything)
#Cell 26 — Collect inputs
name = input("Enter your name: ")
age = int(input("Enter your age: "))
has_ticket = input("Do you have a ticket? (yes/no): ")
```

Check eligibility

```python
#Cell 27 — Check eligibility
is_eligible = age >= 18 and has_ticket.lower() == "yes"
```

Print result




```python
#Cell 28 — Print result
print("--- Result ---")
print("Name:", name)
print("Age:", age)
print("Eligible to enter:", is_eligible)
```
